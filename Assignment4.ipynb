```python
# Titanic Dataset EDA | Google Colab Notebook

# STEP 1: Import required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Set seaborn style
sns.set(style="whitegrid")
plt.rcParams["figure.figsize"] = (10, 6)

# STEP 2: Load the Dataset (upload your titanic_dataset.csv)
from google.colab import files
uploaded = files.upload()

# Load the dataset (fixed filename)
df = pd.read_csv("titanic_dataset.csv")

# STEP 3: Preview the Data
print("Dataset Preview:")
display(df.head())
print("\nData Info:")
df.info()
print("\nSummary Statistics:")
display(df.describe(include='all'))

# STEP 4: Check Missing Values
print("\nMissing Values:")
missing = df.isnull().sum()
print(missing[missing > 0])

# Heatmap of Missing Data
plt.figure(figsize=(10, 6))
sns.heatmap(df.isnull(), cbar=False, cmap="mako")
plt.title("Missing Values Heatmap")
plt.show()

# STEP 5: Univariate Analysis - Distribution of Age
plt.figure()
sns.histplot(df['age'].dropna(), kde=True, bins=30, color='skyblue')
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Count")
plt.show()

# STEP 6: Box Plot for Fare (Outlier Detection)
plt.figure()
sns.boxplot(x=df['fare'], color='salmon')
plt.title("Fare Boxplot (Outlier Detection)")
plt.show()

# STEP 7: Correlation Matrix
plt.figure()
corr = df.select_dtypes(include='number').corr()
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Heatmap")
plt.show()

# STEP 8: Categorical Analysis - Sex vs Survival
plt.figure()
sns.countplot(data=df, x='sex', hue='survived', palette='Set2')
plt.title("Survival Count by Sex")
plt.xlabel("Sex")
plt.ylabel("Count")
plt.legend(title='Survived', labels=['No', 'Yes'])
plt.show()

# STEP 9: Categorical Analysis - Pclass vs Survival
plt.figure()
sns.countplot(data=df, x='pclass', hue='survived', palette='Set3')
plt.title("Survival Count by Passenger Class")
plt.xlabel("Passenger Class")
plt.ylabel("Count")
plt.legend(title='Survived', labels=['No', 'Yes'])
plt.show()

# STEP 10: Violin Plot - Age vs Survival
plt.figure()
sns.violinplot(x='survived', y='age', data=df, palette='pastel')
plt.title("Age Distribution by Survival")
plt.xlabel("Survived")
plt.ylabel("Age")
plt.show()

# STEP 11: Embarked vs Survival
plt.figure()
sns.countplot(data=df, x='embarked', hue='survived', palette='Accent')
plt.title("Survival by Port of Embarkation")
plt.show()

# STEP 12: Clean Summary
print("\nEDA Completed.")
print("Key Insights:")
print("- Survival rate is higher for females and 1st class passengers.")
print("- Fare and age have outliers.")
print("- Strong correlation between fare, class, and survival.")
```
